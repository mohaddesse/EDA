## feature engineering process

for date

```python
df["Date_of_Journey"].str.split("/").str[0]
```

```python
df["Date"]=df["Date_of_Journey"].str.split("/").str[0].astype(int)# df["Date_of_Journey"].apply(lambda x: x.split("/")[0])
df["Month"]=df["Date_of_Journey"].str.split("/").str[1].astype(int)
df["year"]=df["Date_of_Journey"].str.split("/").str[2].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Date_of_Journey"],axis=1,inplace=True)
df.head()
```

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(lambda x: x.split(" ")[0])
```

```python
df.isnull().sum()
```

```python
df["Arrival_hour"]=df["Arrival_Time"].apply(lambda x: x.split(":")[0]).astype(int)
df["Arrival_min"]=df["Arrival_Time"].apply(lambda x: x.split(":")[1]).astype(int)
df.head()
```

```python
df.info()
```

```python
df.drop(["Arrival_Time"],axis=1,inplace=True)
```

```python
df
```

```python
df["Dep_hour"]=df["Dep_Time"].apply(lambda x: x.split(":")[0]).astype(int)
df["Dep_min"]=df["Dep_Time"].apply(lambda x: x.split(":")[1]).astype(int)
df.head()
```

```python
df.drop(["Dep_Time"],axis=1,inplace=True)
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].isnull().sum()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, "nan":1, '4 stops':4})
```

```python
df
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['Total_Stops'])
```

```python
df.isnull().sum()
```

```python
df.drop(["Route"],axis=1,inplace=True)
df.head()
```

```python
df["Additional_Info"].unique()
```

```python
df["hour"]=df["Duration"].str.split(" ").str[0].str.strip("h").astype(int)
df["min"]
```

```python
x=df["Duration"].str.split(" ")
h=[]
m=[]
for i in x:
    if len(i)==2:
        h.append(int(i[0].strip("h")))
        m.append(int(i[1].strip("m")))
    else:
        h.append(int(i[0].strip("h")))
        m.append(0)
        
```

```python
df["duration"]=np.array(h)*60+np.array(m)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df["Airline"]=label.fit_transform(df["Airline"])
df["Source"]=label.fit_transform(df["Source"])
df["Destination"]=label.fit_transform(df["Destination"])
df["Additional_Info"]=label.fit_transform(df["Additional_Info"])
```
