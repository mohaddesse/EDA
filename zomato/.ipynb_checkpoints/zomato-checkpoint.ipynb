# zomato dataset exploratory data analysis (EDA)

<b> Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### In EDA what do we do??
1. missing values
2. explore about the numerical variables
3. explore about the categorical variables
4. finding relationships between features


```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum()]
```

```python
df.shape
```

```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df["Country Code"].head()
```

```python
final_df=pd.merge(df,df_country,how="left",on="Country Code")
final_df.head()
```

```python
## to check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
# pie chart
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

### so zomato maximum records or transication are from india after that US and then united kingdom

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(x=ratings["Aggregate rating"],y=ratings["Rating Count"],hue=ratings["Rating color"],palette=["gray","red","orange","yellow","green","green"])
```

<b>1. not rated count is very high
    <br>
2. maximum number of rating are between 2.8 to 3.9 </b>

```python
sns.countplot(x=ratings["Rating color"],palette=["gray","red","orange","yellow","green","black"])
```
